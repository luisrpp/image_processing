```ruby
require 'image_processing'
```

```ruby
image = Vips::Image.new_from_file('../../samples/apple_2.jpg')
image.display
```

```ruby
grey = image.to_greyscale
(grey > grey.max / 1.1).display
```

```ruby
dilate = ImageProcessing::Morphology.dilate(grey, 5)
dilate.display
```

```ruby
erode = ImageProcessing::Morphology.erode(grey, 5)
erode.display
```

```ruby
gradient = ImageProcessing::Morphology.gradient(grey)
gradient.display
```

```ruby
filtered_gradient = gradient.threshold(gradient.max / 5)
filtered_gradient.display
```

```ruby
hough_transform = ImageProcessing::HoughTransform.new(gradient)

lines = hough_transform.find_lines(theta_res: Math::PI / 360, rho_res: 0.25, threshold: 60)

lines.each { |line| IRuby.display(line) }; puts ''
```

```ruby
hough_transform.accumulator_image.display
```

```ruby
result = image.draw_lines([0.0, 0.0, 255.0], lines)
result.display
```
